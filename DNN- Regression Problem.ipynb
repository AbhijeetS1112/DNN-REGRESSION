```python
# Data Preprocessing
# Only Missing Value
# Split The data Into Dependent And Independent variable
x = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
```

```python
pd.DataFrame(x)
```

```python
# Split The Into Training(For Model Building) And Test(Model Evaluation)
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.75,random_state=101)
```

# Building Deep Neural Network

```python
from tensorflow import keras
```

```python
# Architecture
dnn=tf.keras.models.Sequential()
```

```python
# Adding first Hidden Layer
dnn.add(tf.keras.layers.Dense(units=6, activation="relu"))
```

```python
# Adding Second hidden layer
dnn.add(tf.keras.layers.Dense(units=6,activation="relu"))
```

```python
# Adding Output Layer
dnn.add(tf.keras.layers.Dense(units=1))
```

```python
# Training The Deep Neural Network
dnn.compile(optimizer="adam",loss = "mean_squared_error")
```

```python
# Test The Model
dnn.fit(x_train,y_train,batch_size=32,epochs=100)
```
